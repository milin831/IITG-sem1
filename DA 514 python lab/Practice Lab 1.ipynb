```python
# None, int, float, string, bool

a = 2
b = 2.5
c = "Hi"
d = True
e = None
print(type(a), type(b), type(c), type(d), type(e))  # type() function can work with variables
print(type(2), type(2.5), type("Hi"), type(True), type(None))  # and also with constants
```

```python
# type casting - str(), int(), float(), bool()
a1 = float(a)
print(type(a1))

b1 = int(b)
print(type(b))

d1 = int(d)
print(type(d1))

e1 = bool(e)
print(type(e1))
```

```python
# escap characters
a = "abc\ndef"      # newline escap character
print(type(a), a)

print("ghi\tjkl")   # tab escap character
```

```python
# for
a = list(range(10))
print(a, type(a))

for i in a:
    print(i, end=" ")
```

```python
# while
i = 0
while i<10:
    print(i, end=" ")
    i+=1
```

```python
# pass
# break
# continue

a = list(range(10))
print(a, type(a))

for i in a:
    if (i == 3):
        pass
    elif (i > 6 and i < 8):
        continue
    elif (i==9):
        break
    
    print(i, end=" ")
```

```python
# pass
# break
# continue

i = -1
while (i<10):
    i+=1
    if (i == 3):
        pass
    elif (i > 6 and i < 8):
        continue
    elif (i==9):
        break
    
    print(i, end=" ")
```
