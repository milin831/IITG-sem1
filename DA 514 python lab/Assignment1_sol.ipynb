```python


year = 1947
day = 227
weekday = 5

# upper bound 213 day of year 2025
 
sundays = 0

for i in range(1947,2026):

    year_days = 365

    # leap year
    if i%400==0 or (i%4==0 and i%100!=0):
        year_days=366
 
    for j in range(day,year_days+1):
        day =(day+1)%year_days
        weekday =(weekday + 1)%7

        if weekday==0 :
            sundays+=1

        if year ==2025 and day>213 : 
            break   
    
    year+=1
    
    if year ==2025 and day>213 : 
        break
 
print(sundays)
```
