```python
"""Write a Python code that takes a user-provided sentence as an input and displays a list containing all
the words in the sentence, preserving their original order in the sentence, including any duplicates without using splits."""

sentence = input("Enter a sentence: ")
words = []
word = ""
for ch in sentence:
   if ch != " ":
       word += ch
   else:
       words.append(word)
       word = ""
# Append last word
if word != "":
   words.append(word)
print("Input sentence:", sentence)
print("List of words:", words)
```

```python
"""Write a Python code that takes a user-provided sentence as input and counts the frequency of each
word in the sentence without using any inbuilt function. Display a dictionary where the keys are the
words, and the values are their respective frequencies."""


sentence = input("Enter a sentence: ")

# Step 1: Split into words without split()

words = []

word = ""

for ch in sentence:

    if ch != " ":

        word += ch

    else:

        words.append(word)

        word = ""

if word != "":

    words.append(word)

# Step 2: Count frequencies manually

freq_dict = {}

for w in words:

    if w in freq_dict:

        freq_dict[w] += 1

    else:

        freq_dict[w] = 1

# Step 3: Sort by frequency (highest to lowest) manually

sorted_words = []

temp_dict = freq_dict.copy()

while len(temp_dict) > 0:

    max_word = None

    max_count = -1

    for k in temp_dict:

        if temp_dict[k] > max_count:

            max_count = temp_dict[k]

            max_word = k

    sorted_words.append(max_word)

    del temp_dict[max_word]

print("Input sentence:", sentence)

print("Word frequencies:", freq_dict)

print("Sorted words by frequency:", sorted_words)
```

```python
""" Write a Python program that takes a paragraph as input and counts the number of vowels in each
sentence. Create a dictionary where each key is a separate sentence, and the corresponding value is
another dictionary. In this inner dictionary, each key is a vowel present in the sentence, and its value
is a list of words from the sentence that contain that vowel. Process this dictionary and display the
output along with the paragraph as shown. """

paragraph = input("Enter a paragraph: ")

# Step 1: Split paragraph into sentences manually

sentences = []

sentence = ""

for ch in paragraph:

    sentence += ch

    if ch in ".!?":

        sentences.append(sentence.strip())

        sentence = ""

if sentence.strip() != "":

    sentences.append(sentence.strip())

# Step 2: For each sentence, map vowels to words containing them

vowels = ['a', 'e', 'i', 'o', 'u']

for sent in sentences:

    print(sent)

    vowel_dict = {}

    # Split sentence into words manually

    words = []

    word = ""

    for ch in sent:

        if ch != " ":

            word += ch

        else:

            words.append(word.strip(".,!?"))

            word = ""

    if word != "":

        words.append(word.strip(".,!?"))

    # Create vowel mapping

    for v in vowels:

        words_with_vowel = []

        for w in words:

            lw = w.lower()

            for ch in lw:

                if ch == v:

                    if w not in words_with_vowel:

                        words_with_vowel.append(w)

                    break

        if words_with_vowel:

            vowel_dict[v] = words_with_vowel

    # Print vowel dictionary

    for v in vowel_dict:

        print(f"{v} : {vowel_dict[v]}")
 
```
