### itterating file directory

```python
from pathlib import Path

# creates a Path object that represents a directory path
p = Path(r"D:\OneDrive - Indian Institute of Technology Guwahati\TA DUTIES\DA514 Python Programming Lab (July-Dec, 2025)")
print(type(p))
paths = [x for x in p.iterdir() if x.is_dir()]  # iterdir() itterates over all entries and is_dir() checks if its a directory

for i in paths:
    print(i)
```

### glob() function with wildcards patterns

```python
print(p)
print(type(p))
# glob() search for all path names inside p matching the pattern with wildcards
paths = list(p.glob("**/*.ipynb"))  # '*.py' search for all file name ending with .py and **/ means look inside subdirectories

print(len(paths))
for i in paths:
    print(i)
```

### opening and reading a file

```python
# p = Path(os.path.join(home_path, "Assignment 7 & 8", "test.txt"))
# print(p, type(p))

# 'r' : open for reading (default)
# 'w' : open for writing, truncating the file first
# 'x' : open for exclusive creation, failing if the file already exists
# 'a' : open for writing, appending to the end of file if it exists
# 'b' : binary mode
# 't' : text mode (default)
# '+' : open for updating (reading and writing)

p = Path(r"D:\OneDrive - Indian Institute of Technology Guwahati\TA DUTIES\DA514 Python Programming Lab (July-Dec, 2025)\Assignment 7 & 8\test.txt")
print(p)
if p.exists():
    with p.open('w') as f:
        # line = f.readlines()    # readline() to read one line and read(1) to read one character
        p.write_text('newly written again')
        # print(line)

    with p.open('r') as f:
        line = f.readlines()    # readline() to read one line
        print(line)
else:
    print("p dosen't exists")
```

### jpeg image

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img = mpimg.imread("example_image.jpeg")
plt.imshow(img)
plt.axis("off")   # hide axes

p = Path("example_image.jpeg")
print(p.suffix)

print(type(img), img.shape) # height, width, channels

print(img[0 , 0])
print(img[500 , 500])

first_channel = img[:, :, 0]
print(type(first_channel), first_channel.shape, first_channel[0, 0])
print(np.max(first_channel, axis=(0, 1)))
print(np.min(first_channel, axis=(0, 1)))
```

### creating a dummy image

```python
dummy_image = np.zeros((5, 5, 3), dtype=np.uint8)
print(type(dummy_image), dummy_image.shape)

for i in range(5):
    for j in range(5):
        dummy_image[i, j] = [i*50, j*50, (i+j)*25]  # [R, G, B] values

# Print the array to see pixel values
print("Pixel values:\n", dummy_image)

# Display the image
plt.imshow(dummy_image)
plt.title("Dummy Image")
plt.axis("off")  # hide axes
plt.show()
```
