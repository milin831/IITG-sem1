```python
# list.copy(), [:]
b = a.copy()
print(a, id(a), b, id(b))
```

### dictionaries

```python
# del dict[key]
a = {1: 20, "abc": 30}
print(a)
del a[1]
print(a)
```

```python
# dict[key] = value, key should be unique and key should be immutable
print(a["abc"])
```

```python
# dict.clear(), removes all key-value pairs
a.clear()
print(a)
```

```python
# dict.copy()
```

```python
# dict.get(key)
a[10]= "abc"
a[20]= "def"
print(a)
a.get(10)
```

```python
# dict.items()
print(a.items())
```

```python
# dict.keys()
print(a.keys(), type(a.keys()))
```

```python
# dict.values()
print(a.values(), type(a.values()))
```

```python
# dict will not maintain order
```

```python
# itterating over a dict, itterates over keys
for i in a:
    print(i)
```

```python
# itterating over key-value pair with for k, v in dict.items()
for k, v in a.items():
    print(k, v)

print(a)
```
