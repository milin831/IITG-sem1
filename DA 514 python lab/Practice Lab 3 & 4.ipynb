```python
# parameters data type known at runtime, dynamic typing
def display(a, b):
    print("hello", a, b)
    print(type(a), type(b))

display(2, "Bipul")
display(5, 2.5)
```

```python
# return statement, not mandatory, multiple return elements retuned as tuple
def sum(a, b):
    c = a + b
    return a, b, c

f = sum(2, 5)
print(f, type(f))
```

```python
# function is an object, use with/without brackets ()
def sum(a, b):
    c = a + b
    return a, b, c

f = sum(2, 5)
print(f, type(f))

e = sum
print(type(e), e(2, 5))
print(id(e), id(sum))
```

```python
# positonal arguments, number of arguments and their positon should align
def sum(a, b):
    c = a + b
    return a, b, c

f = sum(2, 5)
print(f, type(f))
```

```python
# keyword arguments, number should match but order can be different
def sum(a, b):
    print(a, b)

sum(b=5, a=2)
```

```python
# default arguments, mapping starts from left
def sum(a, b=5):
    print(a, b)

sum(2, 3)
sum(2)
```

```python
# variagle length arguments, starts with a star * before it in the function definition, stores as tuple
def sum(a, *b):
    print(a, b)

sum(2, 3, 5)
```

```python
# keyword variable length arguments, starts with double star **, stores as dictionary
def sum(a, **b):
    print(a, b)

sum(2, c=3, d=8)
```

```python
# local and global scope, scope starts from where variable is defined, by default local scope is prefered
def sum(a, b=5):
    c = a + b
    print(a, b, c)
    return c

d = sum(2, 3)
print(d)
```

### Classes and Objects

```python
# class definition, creating object
class demo:
    def display(self):
        print("Hello")

test = demo()
print(type(test))

test.display()
```

```python
# Function and Method are same except their placement and the way they are called.
# Function called with just name, Method called with Object/Class name with dot operator 
```

```python
# constructor, self keyword, cls keyword
class demo:
    def __init__(self):
        print("constructor")
        
    def display(self):
        print("Hello")

test = demo()
print(type(test))

test.display()
```

```python
# instance and class variables
class demo:
    c = 30

    def __init__(self):
        self.a = 10
        print("constructor")
        
    def display(self):
        self.b = 20
        print("Hello")

test = demo()
test.display()
```

```python
# instance, class 
class demo:
    c = 30

    def __init__(self):
        self.a = 10
        print("Inside Constructor")
        
    def display(self):
        self.b = 20
        print("printing c", self.c)
        print("Inside Instance Method")
    
    @classmethod
    def class_display(cls):
        cls.d = 40
        print("inside class method")
        print("Inside Class Method")

test = demo()
test.display()
test.class_display()
demo.class_display()
demo.c = 100
print(demo.d)
print(demo.c)
```

```python
# inheritence, single and multiple, super()
class a:
    def __init__(self):
        print("A constructor")

    def display(self):
        print("a instance method")

class c:
    def __init__(self):
        print("C constructor")

class b(c, a):
    def __init__(self):
        super().__init__()
        self.a = 10
        print("B constructor")

    def display_b(self):
        super().display()
        print("b instance method")


test = b()
test.display_b()
test.display()
```

```python
class A:
    def __init__(self):
        print("A constructor")
        super().__init__()   # ensures chain continues

class B:
    def __init__(self):
        print("B constructor")
        super().__init__()

class C(B, A):   # note order: A before B
    def __init__(self):
        print("C constructor")
        super().__init__()

test = C()
```
