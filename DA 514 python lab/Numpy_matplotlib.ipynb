# Numpy

### Import the NumPy library with alias np


```python
import  numpy as np
```

```python
# Create a 1-dimensional NumPy array 
arr=np.array(range(1,10,2))
arr

# Create a 2-dimensional NumPy array using np.array()
arr1=np.array([[1,2],[3,4]])
arr1
# Create a 3-dimensional Numpy array using np.array()
```

```python
# Attributes of  array
np.shape(arr1)
np.ndim(arr1)
arr.dtype
```

```python
# Create an array of shape (10, 2) with zeros
np.zeros([10,2])
np.ones([5,5])*5
```

```python
# Create an array of shape (7, 2, 3) of only fives
np.ones([7,2,3])*5
```

```python
# Create an array within a range of 0 and 100 with step size 4
np.arange(0,100,4)
```

```python
# Create a random array with numbers between 0 and 10 of size (6, 2)
np.random.randint(0,10,(6,2))
```

```python
# Create a random array of floats between 0 & 1 of shape (4, 5)
np.random.uniform(0,1,(5,2))
```

```python
# Set the random seed to 42
np.random.seed(123)
# Create a random array of numbers between 0 & 10 of size (4, 6)
np.random.uniform(0,1,(5,2))
```

```python
# Create an array of random numbers between 1 & 10 of size (3, 7)
arr=np.random.randint(1,10,(3,7))
print(arr)
# Find the unique numbers in the array you just created
print(np.unique(arr))
```

```python
# Find the 0'th index of the latest array you created
arr[0][0]
```

```python
# Get the first 2 rows of latest array you created
arr[:2]
```

```python
# Get the first 2 values of the first 2 rows of the latest array
arr[:2,:2]
```

```python
# algebric operations and reshape()
arr1=np.array([[1,2],[3,4],[5,6]])
arr2=np.reshape(arr1,(2,3)) 
print(arr2)       
arr2@arr
```

```python
# numpy string operations
a="hey guys i'm trying to teach you"
np.char.split(a,sep=" ",maxsplit=1)
```
