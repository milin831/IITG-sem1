```python
# Task 2 
sentence = input("Enter a sentence: ")

# Extract words manually
words = []
word = ""
for ch in sentence:
    if ch != " ":
        word += ch
    else:
        if word != "":
            words.append(word)
            word = ""
if word != "":
    words.append(word)

# Count frequencies
freq = {}
for w in words:
    found = False
    for key in freq:
        if key == w:
            freq[key] += 1
            found = True
            break
    if not found:
        freq[w] = 1

# Manual sort by frequency (highest to lowest)
sorted_words = []
freq_copy = {}
for k in freq:
    freq_copy[k] = freq[k]

while len(freq_copy) > 0:
    max_word = None
    max_count = -1
    for k in freq_copy:
        if freq_copy[k] > max_count:
            max_count = freq_copy[k]
            max_word = k
    sorted_words.append(max_word)
    del freq_copy[max_word]

print(sentence)
print(freq)
print(sorted_words)
```

```python
# Task 3 
paragraph = "Artificial Intelligence is revolutionizing the world. It affects every sector! Will it replace humans?"

# Separate sentences manually
sentences = []
temp = ""
for ch in paragraph:
    temp += ch
    if ch in ".!?":
        if len(temp) > 0:
            # Remove leading/trailing spaces manually
            start = 0
            while start < len(temp) and temp[start] == " ":
                start += 1
            end = len(temp) - 1
            while end >= 0 and temp[end] == " ":
                end -= 1
            cleaned = ""
            for i in range(start, end + 1):
                cleaned += temp[i]
            sentences.append(cleaned)
        temp = ""

vowels = ['a', 'e', 'i', 'o', 'u']

print(paragraph)
print()

for sent in sentences:
    print(sent)
    # Convert to lowercase manually & remove punctuation
    sent_lower = ""
    for ch in sent:
        if ch.isalpha() or ch == " ":
            if 'A' <= ch <= 'Z':
                sent_lower += chr(ord(ch) + 32)
            else:
                sent_lower += ch
        else:
            sent_lower += " "

    # Extract words manually
    words = []
    word = ""
    for ch in sent_lower:
        if ch != " ":
            word += ch
        else:
            if len(word) > 0:
                words.append(word)
                word = ""
    if len(word) > 0:
        words.append(word)

    # Build vowel dictionary without duplicates
    vowel_dict = {}
    for w in words:
        for v in vowels:
            for ch in w:
                if ch == v:
                    if v not in vowel_dict:
                        vowel_dict[v] = []
                    # Check duplicate manually
                    found = False
                    for existing in vowel_dict[v]:
                        if existing == w:
                            found = True
                            break
                    if not found:
                        vowel_dict[v].append(w)
                    break

    # Print vowel dictionary
    for v in vowel_dict:
        print(f"{v} : {vowel_dict[v]}")
    print()
```
